```python
# import libraries
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
# read in data and store in data frame variable
df = pd.read_csv('kiva_data.csv')
# check import
#print(df.head())
```

```python
# create bar chart using Seaborn comparing loan amount by country and gender
sns.set_style('whitegrid')
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(data=df, x = 'country', y = 'loan_amount', hue = 'gender', palette = 'Accent')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set_title('Average Loan Amount Awarded by KIVA Based on Country and Gender')
sns.set_context('notebook', font_scale=1.25)

plt.show()
plt.clf()
```

```python
# create box plot using Seaborn to visualize distribution of loan amounts per country
f, ax = plt.subplots(figsize=(16, 10))
sns.boxplot(data=df, x='country', y='loan_amount', palette='Set1')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set_title('Distribution of KIVA Loan Amounts Based on Country')
sns.set_context('notebook', font_scale=1.25)

plt.show()
```

```python
# create box plot using Seaborn to visualize distribution of loan amounts based on activity
f, ax = plt.subplots(figsize=(16, 10))
sns.boxplot(data=df, x='activity', y='loan_amount', palette='Set2')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set_title('Distribution of KIVA Loan Amounts Based on Activity')
sns.set_context('notebook', font_scale=1.25)

plt.show()
```

```python
# create violin plot using Seaborn to visualize the KDE distributions of loans based on activity
f, ax = plt.subplots(figsize=(16, 10))
sns.violinplot(data=df, x="activity", y="loan_amount", palette='Set3')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set_title('KIVA Loan Amount Distribution by Activity')
sns.set_context('notebook', font_scale=1.25)

plt.show()
```

```python
# create violin plot using Seaborn to visualize the KDE distributions of loans based on country
f, ax = plt.subplots(figsize=(16, 10))
sns.violinplot(data=df, x='country', y='loan_amount', palette='Set3')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set_title('KIVA Loan Amount Distribution by Country')
sns.set_context('notebook', font_scale=1.25)

plt.show()
```

```python
# create violin plot using Seaborn to visualize the KDE distributions of loans based on country and gender
sns.set_palette("Spectral") ## another way to set palette
f, ax = plt.subplots(figsize=(18, 12))
sns.violinplot(data=df, x='country', y='loan_amount', hue='gender')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.set_title('KIVA Loan Amount Distribution by Country and Gender')
sns.set_context('notebook', font_scale=1.25)

plt.show()
```
